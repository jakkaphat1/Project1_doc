# Plagiarism Process

```python
from google.colab import files

def upload_pdfs():
    uploaded = files.upload()
    pdf_paths = list(uploaded.keys())
    return pdf_paths

pdf_paths = upload_pdfs()
```

```python
import fitz

def extract_text_and_embed(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text("text") + "\n"

    print("ข้อความต้นฉบับ :")
    print(text)

    #pythainlp
    tokens = word_tokenize(text, engine="newmm")
    tokenized_text = " ".join(tokens)

    print("ตัดคำ :")
    print(tokenized_text)

    text_embedding = labse_model.encode(tokenized_text)

    return text.strip(), text_embedding


def extract_images_and_embed(pdf_path):
    doc = fitz.open(pdf_path)
    image_embeddings = []
    temp_dir = tempfile.mkdtemp()

    for page_num in range(len(doc)):
        page = doc[page_num]
        image_list = page.get_images(full=True)

        for img_index, img_info in enumerate(image_list):
            xref = img_info[0]
            base_image = doc.extract_image(xref)
            img_bytes = base_image["image"]

            # บันทึกไฟล์รูปภาพ
            img_path = os.path.join(temp_dir, f"page_{page_num}_img_{img_index}.png")
            with open(img_path, "wb") as f:
                f.write(img_bytes)

            # โหลดรูปภาพและสร้าง embedding
            img = Image.open(img_path).convert("RGB")
            inputs = clip_processor(images=img, return_tensors="pt")
            with torch.no_grad():
                img_embedding = clip_model.get_image_features(**inputs)

            image_embeddings.append(img_embedding.squeeze().numpy())

    # ลบไฟล์ชั่วคราว
    import shutil
    shutil.rmtree(temp_dir)

    return image_embeddings
```

```python
from slugify import slugify
import numpy as np

def to_ascii_id(filename):
    return slugify(filename, separator="_")
```

```python
def upload_text_embedding(doc_id, text_embedding):
    if text_embedding is None or len(text_embedding) == 0:
        print(f"ไม่มีข้อความ")
        return
    text_index.upsert([(f"text_{doc_id}", text_embedding.tolist(), {"type": "text"})])
    print(f"อัปโหลด Text Embedding ของ{doc_id} ได้")

def upload_image_embedding(doc_id, image_embeddings):
    if image_embeddings is None or len(image_embeddings) == 0:
        print(f"ไม่มีรูปภาพใน {doc_id}")
        return
    for i, img_emb in enumerate(image_embeddings):
        image_index.upsert([(f"image_{doc_id}_{i}", img_emb.tolist(), {"type": "image"})])
    print(f"อัปโหลด Image Embedding ของ{doc_id} ได้")
```

```python
for pdf_path in pdf_paths:
    doc_id = to_ascii_id(os.path.basename(pdf_path).split('.')[0])  # เปลี่ยนเป็น ASCII
    text, text_embedding = extract_text_and_embed(pdf_path)
    image_embeddings = extract_images_and_embed(pdf_path)

    upload_text_embedding(doc_id, text_embedding)
    upload_image_embedding(doc_id, image_embeddings)
```

```python
def compare_images(image_embeddings, doc_id_1, doc_id_2):
    """
    เปรียบเทียบทุกรูปภาพของ doc_1 และ doc_2 แล้วหาค่าความคล้ายกันสูงสุด
    """
    max_similarity = 0.0
    image_keys_1 = [key for key in image_embeddings.keys() if key.startswith(f"image_{doc_id_1}_")]
    image_keys_2 = [key for key in image_embeddings.keys() if key.startswith(f"image_{doc_id_2}_")]

    if not image_keys_1 or not image_keys_2:
        return 0.0

    for img_1 in image_keys_1:
        for img_2 in image_keys_2:
            similarity = cosine_similarity(
                [image_embeddings[img_1]], [image_embeddings[img_2]]
            )[0][0]
            max_similarity = max(max_similarity, similarity)
    return max_similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_all_text_embeddings():
    text_vectors = text_index.fetch(ids=[f"text_{to_ascii_id(os.path.basename(pdf).split('.')[0])}" for pdf in pdf_paths])
    return {k: np.array(v.values) for k, v in text_vectors.vectors.items()}

def get_all_image_embeddings():
    stats = image_index.describe_index_stats()
    total_vectors = stats["total_vector_count"]

    if total_vectors == 0:
        print("ไม่มีรูปภาพใน index")
        return {}

    #ใช้ query() เพื่อดึงค่าเวกเตอร์ทั้งหมดแทน fetch()
    image_vectors = image_index.query(
        vector=[0] * 768,
        top_k=total_vectors,
        include_values=True
    )

    return {match["id"]: np.array(match["values"]) for match in image_vectors["matches"]}
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def compare_all_documents(top_k=5):
    text_embeddings = get_all_text_embeddings()
    image_embeddings = get_all_image_embeddings()

    if not text_embeddings and not image_embeddings:
        print("ไม่มีข้อมูลใน Pinecone")
        return []

    results = []

    for doc_id in text_embeddings.keys():
        similar_docs = text_index.query(
            vector=text_embeddings[doc_id].tolist(),
            top_k=top_k,
            include_metadata=True
        )

        for match in similar_docs["matches"]:
            doc_id_2 = match["id"]
            if doc_id == doc_id_2:
                continue

            text_similarity = match["score"]
            image_similarity = compare_images(
                image_embeddings,
                doc_id.replace("text_", ""),
                doc_id_2.replace("text_", "")
            )
            if text_similarity >= 0.90:
                is_plagiarized = True
            elif text_similarity >= 0.85 and image_similarity >= 0.85:
                is_plagiarized = True
            else:
                is_plagiarized = False

            results.append({
                "doc_1": doc_id,
                "doc_2": doc_id_2,
                "text_similarity": text_similarity,
                "image_similarity": image_similarity,
                "plagiarized": is_plagiarized
            })

            status = "มีความคล้ายสูง" if is_plagiarized else "มีความคล้ายต่ำ"
            print(f"{doc_id} vs {doc_id_2} ---> Text: {text_similarity:.4f}  ,  Image: {image_similarity:.4f} → {status}")

    return results
```

```python
compare_results = compare_all_documents()
```
